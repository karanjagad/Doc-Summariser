```python
def freq(str): 
    newdict={}
    # break the string into list of words  
    str = str.split()          
    str2 = [] 
  
    # loop till string values present in list str 
    for i in str:              
  
        # checking for the duplicacy 
        if i not in str2: 
  
            # insert value in str2 
            str2.append(i)  
              
    for i in range(0, len(str2)): 
  
        new = str2[i], ':', str.count(str2[i])
        newdict[str2[i]]= str.count(str2[i])


    
        
    return newdict
        
```

Frequency of words in each sentence: the output should

```python
newaa = {}
for item in a_list:
    newaa.update({item:freq(item)})
# newdict = freq(a_list[0])
print(newaa)
```

```python
def termfreq(freq_matrix):
    tf_matrix = {}

    for sent, f_table in freq.items():
        tf_table = {}

        count_words_in_sentence = len(f_table)
        for word, count in f_table.items():
            tf_table[word] = count / count_words_in_sentence

        tf_matrix[sent] = tf_table

    return tf_matrix
```


Matrix termFrequency: The termFrequency matrix should
be a dictionary where each key is a sentence and the value is also a
dictionary of word frequency.


```python
def _create_documents_per_words(freq_matrix):
    word_per_doc_table = {}

    for sent, f_table in freq_matrix.items():
        for word, count in f_table.items():
            if word in word_per_doc_table:
                word_per_doc_table[word] += 1
            else:
                word_per_doc_table[word] = 1

    return word_per_doc_table
```
